```python
import requests
from bs4 import BeautifulSoup
```

```python
#url = "https://wuzzuf.net/search/jobs?q=data&start=1"
def scrape (page_num):
  url = f"https://wuzzuf.net/search/jobs?q=data&start={page_num}"
  response = requests.get(url)
  soup = BeautifulSoup(response.text)
  blocks = soup.find_all("div",{'class' : 'css-ghe2tq e1v1l3u10'})
  jobs = []

  for block in blocks:
    jtitle = block.find_all("h2",{'class' : 'css-193uk2c'})
    job_title = jtitle[0].text.strip() if jtitle else "N/A"

    ncompany = block.find_all("span",{'class' : 'css-16x61xq'})
    company_name = ncompany[0].text.strip() if ncompany else "N/A"

    jtype = block.find_all("span",{'class' : 'css-uc9rga eoyjyou0'})
    job_type = jtype[0].text.strip() if jtype else "N/A"

    tjobs = block.find_all("span",{'class' : 'css-uofntu eoyjyou0'})
    job_time = tjobs[0].text.strip() if tjobs else "N/A"

    jobs.append({
      'job_title': job_title,
      'company_name': company_name,
      'job_type': job_type,
      'work_mode':job_time
    })

  return jobs
```

```python
import tqdm
all_jobs = []
for i in tqdm.tqdm(range(0,100)):
  jobs = scrape(i)
  all_jobs.extend(jobs)
```
