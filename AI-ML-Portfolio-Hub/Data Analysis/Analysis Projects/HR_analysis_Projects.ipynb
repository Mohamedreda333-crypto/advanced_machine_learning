### **Is the employee turnover rate (attrition) high?**

```python
attrition_counts = df['Attrition'].value_counts()
attrition_percent = attrition_counts / len(df) * 100

fig = px.pie(
    names=attrition_counts.index,
    values=attrition_counts.values,
    title="🔹 Employee Attrition Distribution",
    color=attrition_counts.index,
    color_discrete_map={'Yes': '#FF6B6B', 'No': '#4ECDC4'},
    hole=0.45
)

fig.update_traces(textinfo='percent+label', pull=[0.1, 0])
fig.update_layout(
    title_font_size=20,
    title_x=0.5,
    template='plotly_white'
)

fig.show()
```

### **Is there a relationship between an employee's age and the likelihood of them leaving the company (attrition)?**


```python
fig = px.box(
    df,
    x='Attrition',
    y='Age',
    color='Attrition',
    color_discrete_map={'Yes': '#FF6B6B', 'No': '#4ECDC4'},
    title="🔹 Age Distribution by Attrition Status"
)

fig.update_layout(
    template='plotly_white',
    title_font_size=20,
    title_x=0.5,
    xaxis_title="Attrition Status",
    yaxis_title="Age",
    font=dict(size=14)
)

fig.show()
```

### **Does monthly income have an impact on employee turnover (attrition)?**

```python
fig = px.histogram(
    df,
    x='MonthlyIncome',
    color='Attrition',
    nbins=40,
    barmode='overlay',
    opacity=0.7,
    color_discrete_map={'Yes': '#FF6B6B', 'No': '#4ECDC4'},
    title="📊 Monthly Income Comparison by Attrition Status"
)

fig.update_layout(
    template='plotly_white',
    title_font_size=22,
    title_x=0.5,
    xaxis_title="Monthly Income",
    yaxis_title="Count",
    font=dict(size=14),
)
fig.show()
```

### **Does the level of job satisfaction have a relationship with the employee turnover rate?**

### **Does the number of years of experience or length of time at the company (YearsAtCompany) affect the likelihood of an employee leaving?**

```python
fig = px.histogram(
    df,
    x='YearsAtCompany',
    color='Attrition',
    barmode='overlay',
    nbins=20,
    opacity=0.7,
    color_discrete_map={'Yes': '#FF6B6B', 'No': '#4ECDC4'},
    title="🔹 Distribution of Years at Company by Attrition"
)

fig.update_layout(
    template='plotly_white',
    title_font_size=20,
    title_x=0.5,
    xaxis_title="Years at Company",
    yaxis_title="Employee Count",
    font=dict(size=14)
)

fig.show()
```

### **Is there a relationship between job type (JobRole) and employee turnover rate?**

```python
role_attrition = (
    df.groupby('JobRole')['Attrition']
    .value_counts(normalize=True)
    .mul(100)
    .rename('Percentage')
    .reset_index()
)

role_attrition_yes = role_attrition[role_attrition['Attrition'] == 'Yes']

fig = px.bar(
    role_attrition_yes.sort_values(by='Percentage', ascending=True),
    x='Percentage',
    y='JobRole',
    orientation='h',
    color='Percentage',
    color_continuous_scale=['#4ECDC4', '#FF6B6B'],
    title="🔹 Attrition Rate by Job Role (%)"
)

fig.update_layout(
    template='plotly_white',
    title_font_size=20,
    title_x=0.5,
    xaxis_title="Attrition Percentage (%)",
    yaxis_title="Job Role",
    font=dict(size=14)
)

fig.show()
```

### **Do frequent business travel or distance from home to work affect an employee's decision to leave?**

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Attrition by Business Travel", "Attrition by Distance From Home"),
    horizontal_spacing=0.12
)

travel_counts = (
    df.groupby('BusinessTravel')['Attrition']
    .value_counts(normalize=True)
    .mul(100)
    .rename('Percentage')
    .reset_index()
)
travel_yes = travel_counts[travel_counts['Attrition'] == 'Yes']

fig.add_trace(
    go.Bar(
        x=travel_yes['BusinessTravel'],
        y=travel_yes['Percentage'],
        marker_color=['#FF6B6B', '#FFA07A', '#4ECDC4'],
        name="Business Travel"
    ),
    row=1, col=1
)

fig.add_trace(
    go.Box(
        y=df[df['Attrition'] == 'Yes']['DistanceFromHome'],
        name='Yes (Left Company)',
        marker_color='#FF6B6B',
        boxmean='sd'
    ),
    row=1, col=2
)

fig.add_trace(
    go.Box(
        y=df[df['Attrition'] == 'No']['DistanceFromHome'],
        name='No (Still Working)',
        marker_color='#4ECDC4',
        boxmean='sd'
    ),
    row=1, col=2
)

fig.update_layout(
    template='plotly_white',
    title="🔹 Effect of Travel & Distance on Employee Attrition",
    title_font_size=20,
    title_x=0.5,
    showlegend=False,
    font=dict(size=13)
)

fig.update_xaxes(title_text="Business Travel Type", row=1, col=1)
fig.update_yaxes(title_text="Attrition Percentage (%)", row=1, col=1)
fig.update_yaxes(title_text="Distance From Home (KM)", row=1, col=2)

fig.show()
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

categorical_cols = df.select_dtypes('object').columns.tolist()
categorical_cols = [col for col in categorical_cols if col not in ['Over18', 'EmployeeCount', 'StandardHours', 'EmployeeNumber']]

results = []
for col in categorical_cols:
    temp = df.groupby(col)['Attrition'].mean().reset_index()
    temp['AttritionRate(%)'] = temp['Attrition'] * 100
    temp['Feature'] = col
    results.append(temp)

summary = pd.concat(results, ignore_index=True)

feature_importance = summary.groupby("Feature")['AttritionRate(%)'].mean().reset_index()
feature_importance = feature_importance.sort_values(by='AttritionRate(%)', ascending=False)


fig = px.bar(
    feature_importance,
    x='AttritionRate(%)',
    y='Feature',
    orientation='h',
    color='AttritionRate(%)',
    color_continuous_scale='Reds',
    text=feature_importance['AttritionRate(%)'].apply(lambda x: f"{x:.2f}%"),
    title='💼 Factors with Highest Attrition Rates (Categorical Overview)'
)

fig.update_traces(textposition='outside')
fig.update_layout(
    template='plotly_white',
    xaxis_title='Average Attrition Rate (%)',
    yaxis_title='Feature',
    font=dict(size=13),
    height=700,
    title_font=dict(size=22, color='#2C3E50', family='Arial Black'),
    coloraxis_showscale=False
)

fig.show()
```
