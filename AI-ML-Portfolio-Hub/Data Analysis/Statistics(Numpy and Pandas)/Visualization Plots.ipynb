```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd 
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

# Univariate plots

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Sex distribution
plt.figure(figsize=(6, 6))
sns.countplot(x='sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
# BMI distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['bmi'], kde=True)
plt.title('BMI Distribution')
plt.show()
```

```python
# Children count distribution
plt.figure(figsize=(6, 6))
sns.countplot(x='children', data=df)
plt.title('Number of Children')
plt.show()
```

```python
# Smoker distribution
plt.figure(figsize=(6, 6))
sns.countplot(x='smoker', data=df)
plt.title('Smoker Status')
plt.show()
```

```python
# Region distribution
plt.figure(figsize=(6, 6))
sns.countplot(x='region', data=df)
plt.title('Region Distribution')
plt.show()
```

```python
# Charges distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['charges'], kde=True)
plt.title('Charges Distribution')
plt.show()
```

# Bivariate Analysis

```python
# Age vs. Charges (Numerical vs. Numerical)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', data=df)
plt.title('Age vs. Charges')
plt.show()
```

```python
# Smoker vs. Charges (Categorical vs. Numerical)
plt.figure(figsize=(10, 6))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Smoker vs. Charges')
plt.show()
```

```python
# Region vs. BMI (Categorical vs. Numerical)
plt.figure(figsize=(10, 6))
sns.violinplot(x='region', y='bmi', data=df)
plt.title('Region vs. BMI')
plt.show()
```

```python
# Children vs. Charges (Categorical vs. Numerical)
plt.figure(figsize=(10, 6))
sns.barplot(x='children', y='charges', data=df)
plt.title('Children vs. Charges')
plt.show()
```

```python
# BMI vs. Charges (Numerical vs. Numerical)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bmi', y='charges', data=df)
plt.title('BMI vs. Charges')
plt.show()
```

# Trivariate Analysis

```python
# Age vs. Charges with Smoker Status (Numerical vs. Numerical with Categorical)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', hue='smoker', data=df)
plt.title('Age vs. Charges (Colored by Smoker Status)')
plt.show()
```

```python
# BMI vs. Charges with Number of Children (Numerical vs. Numerical with Categorical)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bmi', y='charges', hue='children', data=df, palette='coolwarm', size='children', sizes=(20, 200))
plt.title('BMI vs. Charges (Colored by Number of Children)')
plt.show()
```

```python
# Age vs. BMI with Region (Faceting by Region)
g = sns.FacetGrid(df, col="region", col_wrap=2, height=4)
g.map(sns.scatterplot, 'age', 'bmi')
g.fig.suptitle('Age vs. BMI Across Regions', y=1.05)
plt.show()
```

```python
# Charges vs. BMI with Sex and Smoker Status (Numerical vs. Numerical with Categorical Variables)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bmi', y='charges', hue='sex', style='smoker', data=df)
plt.title('Charges vs. BMI (Colored by Sex, Styled by Smoker Status)')
plt.show()
```

```python
sns.pairplot(df, diag_kind='kde', hue='smoker')
plt.suptitle('Pair Plot of Numerical Features (Colored by Smoker Status)', y=1.02)
plt.show()
```
