### **New Column classification based on BMI degrees**

```python
df["BMI Classification"] = np.nan
```

```python
df.head()
```

```python
df.shape[0]
```

```python
for i in range (df.shape[0]):
    if df["bmi"][i] < 18:
        df["BMI Classification"][i] = "Underweight"
        
    elif df["bmi"][i] < 25:
        df["BMI Classification"][i] = "Normal weight"
    
    elif df["bmi"][i] < 30:
        df["BMI Classification"][i] = "Overweight"
    
    else:
        df["BMI Classification"][i] = "Obese"
```

### **New classification by age BMI smoking**

```python
def classify_risk(row):
    if (row["smoker"] == "yes") or (row["bmi"] >= 30) or (row["age"] > 50):
        return "High Risk"
    
    elif (30 <= row["age"] <= 50) and (25 <= row["bmi"] < 30):
        return "Medium Risk"
    
    else:
        return "Low Risk"
```

```python
df["risk_category"] = df.apply(classify_risk, axis=1)    # rows
```

```python
df[["age", "bmi", "smoker", "risk_category"]]
```
