### **Dimensions**

```python
# 0-D
arr3 = np.array(30) #point
print(arr3)
print(type(arr3))
print(arr3.ndim)
```

```python
# 1-D
arr4 = np.array([1,2,3,4,5,6])
print(arr4)
print(type(arr4))
print(arr4.ndim)
```

```python
# 2-D
arr5 = np.array([[1,2,3],[4,5,6]])
print(arr5)
print(type(arr5))
print(arr5.ndim)
```

```python
# 3-D
arr6 = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]]) 
print(arr6)
print(arr6.ndim)

# 2x2x2 (2 rows , 2 coloums , depth =2)
```

```python
print(arr3.ndim)
print(arr4.ndim)
print(arr5.ndim)
print(arr6.ndim)
```

```python
# 4-D
arr7 = np.array([
    [[[1,2,3],[4,5,6]], [[1,2,3],[4,5,6]]],
    [[[1,2,3],[4,5,6]], [[1,2,3],[4,5,6]]]
])

print(arr7)
print(arr7.ndim)

# 3X3X2X2 (3 rows , 3 coloums , 2 layers(depth) in 3D,2 layers(depth) in 4D )
```

```python
# point in 0-D system
np.array(5).ndim
```

```python
# point in 2-D system
np.array([[5]]).ndim
```

```python
arr = np.array([1,3,7],ndmin= 5)
print(arr)
print(arr.ndim)
# 1x3x1x1x1
```

```python
# Concatenation
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.concatenate((arr1 , arr2))
print(arr)
```

```python
arr1 = np.array([[1,2,3],[1,2,3]])
arr2 = np.array([[4,2,6],[1,2,3]])
arr = np.concatenate((arr1 , arr2))
print(arr)
```

```python
arr1 = np.array([[1,2,3],[1,2,3]])
arr2 = np.array([[4,2,6],[1,2,3]])
arr = np.concatenate((arr1 , arr2),axis =1)
print(arr)
```
