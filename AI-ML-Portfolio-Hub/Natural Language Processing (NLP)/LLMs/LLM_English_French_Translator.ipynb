```python
# Import required classes from Hugging Face Transformers
# Hugging Face model card: https://huggingface.co/facebook/mbart-large-50-many-to-many-mmt

from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
model_name = "facebook/mbart-large-50-many-to-many-mmt"

# Load tokenizer and model
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)

# Set source and target language codes
tokenizer.src_lang = "en_XX"  # English input
target_lang_token = tokenizer.lang_code_to_id["ar_AR"]  # Arabic output
```

```python
def translate_en_to_ar(text):
    """
    Translates a sentence from English to Arabic using the mBART model.
    """
    # Tokenize input English text
    input_ids = tokenizer(text, return_tensors="pt").input_ids

    # Generate translation with forced Arabic language token
    output_ids = model.generate(input_ids, forced_bos_token_id=target_lang_token)

    # Decode output tokens into readable Arabic string
    translation = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return translation
```

```python
# ----------------- Run interaction loop -----------------
print("🔁 English to Arabic Translator (type 'exit' to quit)\n")

while True:
    user_input = input("Enter English sentence: ")
    if user_input.lower() == "exit":
        break

    arabic_translation = translate_en_to_ar(user_input)
    print("→ Arabic:", arabic_translation, "\n")
```
