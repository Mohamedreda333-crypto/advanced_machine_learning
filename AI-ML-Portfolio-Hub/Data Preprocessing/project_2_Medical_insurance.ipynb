# **Handle Outlier**

```python
df.columns
```

```python
num_cols = ["age","bmi"]
plt.Figure(figsize=(10,1))
for i , col in enumerate(num_cols):
    plt.subplot(2 , 1 , i+1)
    sns.boxplot(df[col],orient="h")
    plt.title(f"{col} outlier Detection")
plt.subplots_adjust(hspace = 0.8,wspace = 0.3)
plt.show()
```

```python
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
Lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
Lower_Outliers = df[df[col] < Lower_fence][col].values
Upper_Outliers = df[df[col] > upper_fence][col].values
df[col].replace(Upper_Outliers , upper_fence , inplace = True)
df[col].replace(Lower_Outliers , Lower_fence , inplace = True)
```

```python
plt.Figure(figsize=(10,1))
for i , col in enumerate(num_cols):
    plt.subplot(2 , 1 , i+1)
    sns.boxplot(df[col],orient="h")
    plt.title(f"{col} outlier Detection")
plt.subplots_adjust(hspace = 0.8,wspace = 0.3)
plt.show()
```

### **Check the relationship between each feature and the target**

```python
# age column + Target(charges) : Weak positive correlation
# bmi column + Target(charges) : Non correlation
plt.figure(figsize=(10,4))
for i , col in enumerate(num_cols):
    plt.subplot(1 , 2 , i+1)
    plt.scatter(df[col] , df["charges"])
    plt.xlabel(f"{col}")
    plt.ylabel("charges")
    plt.title(f"Relation between {col} and charges")
plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
#insights:
"""
1) sex (male) --> high charges (Approximately 14,000) above than sex(female)
2) childern (2,3) --> high charges (Approximately 15,000) above than childern (0,1,4,5)
3) smoker(yes) --> very high charges (Approximately 30,000)
4) region(southeast) --> high charges (Approximately 15,000)
"""
plt.Figure(figsize=(10,5))
for i , col in enumerate(cat_cols):
    plt.subplot(2,2,i+1)
    classes = df[cat_cols].nunique()
    palette = sns.color_palette("tab10", len(classes))
    sns.barplot(x= col , y="charges" , data = df , palette = palette)
    plt.xlabel(f"{col}")
    plt.ylabel("charges")
    plt.title(f"Relation between {col} and charges")
plt.subplots_adjust(hspace = 0.8 , wspace=1.5)
plt.show()
    
```

### **Variant of each column**

```python
# insights
"""
1) age column --> approximatly uniform distribution
2) bmi column --> Normal distribution(Mean = Median = Mode)
"""
plt.figure(figsize=(10,4))
for i , col in enumerate(num_cols):
    plt.subplot(1,2,i+1)
    # plt.hist(df[cols],color="red", density= True)
    sns.histplot(x = col, data = df, color="red")
    plt.title(f"{col} distribution")
plt.subplots_adjust(hspace = 0.8 , wspace = 0.3)
plt.show()
```
