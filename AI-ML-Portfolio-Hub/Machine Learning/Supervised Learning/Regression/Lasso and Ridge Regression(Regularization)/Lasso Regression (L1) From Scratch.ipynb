```python
import numpy as np
```

```python
class LassoRegression:
    """
    Lasso Regression implemented from scratch using Gradient Descent.

    Lasso Regression applies L1 regularization which encourages sparsity
    and can perform automatic feature selection.
    """

    def __init__(self, lr=0.01, epochs=1000, alpha=1.0):
        """
        Initialize the Lasso Regression model.

        Parameters
        ----------
        lr : float
            Learning rate for gradient updates.
        epochs : int
            Number of training iterations.
        alpha : float
            Regularization strength (L1 penalty).
        """
        self.lr = lr
        self.epochs = epochs
        self.alpha = alpha
        self.weights = None
        self.bias = 0

    def fit(self, X, y):
        """
        Train the Lasso Regression model using gradient descent.

        Parameters
        ----------
        X : numpy.ndarray
            Feature matrix of shape (n_samples, n_features).
        y : numpy.ndarray
            Target values of shape (n_samples,).

        Returns
        -------
        None
        """
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)

        for _ in range(self.epochs):
            y_pred = np.dot(X, self.weights) + self.bias

            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            dw += self.alpha * np.sign(self.weights)

            db = (1 / n_samples) * np.sum(y_pred - y)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        """
        Predict target values using the trained Lasso model.

        Parameters
        ----------
        X : numpy.ndarray
            Input feature matrix.

        Returns
        -------
        numpy.ndarray
            Predicted target values.
        """
        return np.dot(X, self.weights) + self.bias
```
