```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sample data
x = np.array([1 , 2 , 3 , 4 ])
y = np.array([2 , 2.8 , 3.6 , 4.5 ])
```

```python
# intializaion parameter
theta_0 = 0
theta_1 = 0
alpha = 0.01
num_iterations = 1000
```

```python
#store the SSE value for plot(Visualization)
SSE_values = []
```

```python
# Gradient descent
for i in range(num_iterations):
    # Compute prediction
    # y = w(theta_1)x + b(theta_0)
    
    y_pred = theta_0 + theta_1 * x
    
    # Compute Gradients
    d_theta_0 = -2 * np.sum( y - y_pred)
    d_theta_1  = -2 * np.sum ((y - y_pred)* x)
    
    # Update Parameters
    theta_0 -= alpha * d_theta_0       # new b(intercept or bias)
    theta_1 -= alpha * d_theta_1       # new w (coff or slope)
    
    # compute SSE and compute it
    SSE = np.sum((y - y_pred) ** 2)
    SSE_values.append(SSE)
    
    # print SSE of every 100 iterations
    if (i + 1) % 100 == 0 :
        print(f"Iteration {i + 1} , SSE : {SSE}")
    
```

```python
# plotting SSE values
plt.Figure(figsize=(12,5))

# SSE over iteration
plt.subplot(1,2,1)
plt.plot(range(num_iterations),SSE_values)
plt.xlabel ("Iteration")
plt.ylabel("SSE")
plt.title("SSE Over iteration")
plt.legend()


# plottint the Data points and the final regression line 
plt.subplot(1,2,2)
plt.scatter(x,y,edgecolors="blue")
plt.plot (x, theta_0 + theta_1 * x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line Regression fit")
plt.legend()

plt.tight_layout()
plt.show()
```
