# Hyperopt: Model Selection & Hyperparameter Optimization (Detailed Guide)

##  What is this code?
This code uses **Hyperopt** to automatically:
- Select the best classification model
- Tune its hyperparameters
- Using **Bayesian Optimization (TPE)** instead of GridSearch

Models included:
- Random Forest
- Gradient Boosting
- Support Vector Machine (SVM)

---

##  Why Hyperopt?
Traditional GridSearch is slow and tries **all combinations**.  
Hyperopt is **smart**:
- Tries fewer combinations
- Learns from previous trials
- Reaches better results faster

```python
# Import core Hyperopt functions:
# fmin  -> runs the optimization process
# tpe   -> Tree-structured Parzen Estimator (Bayesian optimization algorithm)
# hp    -> used to define the hyperparameter search space
# Trials-> object that stores results of all optimization trials
from hyperopt import fmin, tpe, hp, Trials


# Objective function that Hyperopt tries to MINIMIZE
# It receives a dictionary of hyperparameters sampled from the search space
def objective(params):
    
    # Extract the model type (Random Forest, Gradient Boosting, or SVM)
    model_type = params['model_type']
    
    # Remove model_type from params so it is NOT passed to the model constructor
    # (models do not accept "model_type" as an argument)
    del params['model_type']

    # Create the appropriate model based on the selected model type
    if model_type == "Random Forest":
        # RandomForestClassifier with dynamically selected hyperparameters
        model = RandomForestClassifier(**params, random_state=42)
    
    elif model_type == "Gradient Boosting":
        # GradientBoostingClassifier with its own hyperparameters
        model = GradientBoostingClassifier(**params, random_state=42)
    
    elif model_type == "SVM":
        # Support Vector Machine classifier
        model = SVC(**params)

    # Train the selected model using the scaled feature matrix
    model.fit(X_scaled, y)

    # Predict on the SAME data (note: this is for demonstration, not production)
    predictions = model.predict(X_scaled)

    # Calculate accuracy as the evaluation metric
    accuracy = accuracy_score(y, predictions)

    # Hyperopt always minimizes the objective function
    # To maximize accuracy, we return the NEGATIVE value
    return -accuracy


# Define the hyperparameter search space
# hp.choice selects ONE option from the list at each trial
space = hp.choice('classifier_type', [
    
    # ---------------- Random Forest Search Space ----------------
    {
        'model_type': 'Random Forest',
        
        # Number of trees in the forest
        'n_estimators': hp.choice('n_estimators', [50, 100, 200]),
        
        # Maximum depth of each tree (None = unlimited depth)
        'max_depth': hp.choice('max_depth', [None, 10, 20]),
        
        # Minimum number of samples required to split a node
        'min_samples_split': hp.choice('min_samples_split', [2, 5, 10])
    },

    # ---------------- Gradient Boosting Search Space ----------------
    {
        'model_type': 'Gradient Boosting',
        
        # Number of boosting stages
        'n_estimators': hp.choice('gb_n_estimators', [50, 100, 200]),
        
        # Learning rate controls how much each tree contributes
        'learning_rate': hp.choice('learning_rate', [0.01, 0.1, 0.2]),
        
        # Maximum depth of individual trees
        'max_depth': hp.choice('gb_max_depth', [3, 5, 10])
    },

    # ---------------- SVM Search Space ----------------
    {
        'model_type': 'SVM',
        
        # Regularization parameter (controls margin softness)
        'C': hp.choice('svm_C', [0.1, 1, 10]),
        
        # Kernel type used by SVM
        'kernel': hp.choice('kernel', ['linear', 'rbf'])
    }
])


# Create a Trials object to store detailed information
# about each optimization attempt
trials = Trials()


# Run the Hyperopt optimization process
best_params = fmin(
    fn=objective,          # Objective function to minimize
    space=space,           # Hyperparameter search space
    algo=tpe.suggest,      # Bayesian optimization algorithm
    max_evals=20,          # Number of trials (experiments)
    trials=trials          # Store results of each trial
)


# Extract the best model type selected by Hyperopt
best_model_type = best_params['classifier_type']

# Retrieve the best accuracy achieved (convert back from negative loss)
best_accuracy = -min(trials.losses())


# Final outputs:
# - best_params     : best hyperparameters found
# - best_model_type : best classifier selected
# - best_accuracy   : highest achieved accuracy
best_params, best_model_type, best_accuracy
```

# **Hyper Optimizer With Cross Validation**

```python
# Import core Hyperopt functions
from hyperopt import fmin, tpe, hp, Trials

# Scikit-learn utilities
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


# Objective function for Hyperopt (to be minimized)
def objective(params):
    
    # Extract model type
    model_type = params['model_type']
    
    # Remove model_type before passing parameters to the model
    del params['model_type']

    # Create model based on selected type
    if model_type == "Random Forest":
        model = RandomForestClassifier(**params, random_state=42)
    
    elif model_type == "Gradient Boosting":
        model = GradientBoostingClassifier(**params, random_state=42)
    
    elif model_type == "SVM":
        model = SVC(**params)

    # Perform K-Fold Cross Validation
    # cv=5 means the data is split into 5 folds
    # scoring='accuracy' ensures accuracy is used as evaluation metric
    cv_accuracy = cross_val_score(
        model,
        X_scaled,
        y,
        cv=5,
        scoring='accuracy'
    ).mean()

    # Hyperopt minimizes the objective function
    # We negate accuracy to maximize it
    return -cv_accuracy


# Define hyperparameter search space
space = hp.choice('classifier_type', [

    # ---------------- Random Forest ----------------
    {
        'model_type': 'Random Forest',
        'n_estimators': hp.choice('n_estimators', [50, 100, 200]),
        'max_depth': hp.choice('max_depth', [None, 10, 20]),
        'min_samples_split': hp.choice('min_samples_split', [2, 5, 10])
    },

    # ---------------- Gradient Boosting ----------------
    {
        'model_type': 'Gradient Boosting',
        'n_estimators': hp.choice('gb_n_estimators', [50, 100, 200]),
        'learning_rate': hp.choice('learning_rate', [0.01, 0.1, 0.2]),
        'max_depth': hp.choice('gb_max_depth', [3, 5, 10])
    },

    # ---------------- SVM ----------------
    {
        'model_type': 'SVM',
        'C': hp.choice('svm_C', [0.1, 1, 10]),
        'kernel': hp.choice('kernel', ['linear', 'rbf'])
    }
])


# Store optimization trials
trials = Trials()

# Run Hyperopt optimization
best_params = fmin(
    fn=objective,
    space=space,
    algo=tpe.suggest,
    max_evals=20,
    trials=trials
)

# Extract best model type index
best_model_index = best_params['classifier_type']

# Best cross-validated accuracy
best_accuracy = -min(trials.losses())

best_params, best_model_index, best_accuracy
```
