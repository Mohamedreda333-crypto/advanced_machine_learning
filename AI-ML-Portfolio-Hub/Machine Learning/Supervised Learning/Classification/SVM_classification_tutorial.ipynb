```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**Sepal length vs Sepal Width (Setosa vs Versicolor)**

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+' , label = iris.target_names[0])
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.',  label = iris.target_names[1])
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'],color="red",marker='*',  label = iris.target_names[2])
plt.legend()
plt.show()
```

**Petal length vs Pepal Width (Setosa vs Versicolor)**

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'],color="red",marker='*')
plt.show()
```

**Train Using Support Vector Machine (SVM)**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC # support vector classifier
model = SVC()
# also : SVR:  support vector regressor
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

**Tune parameters**

# **1. Regularization (C)**

<img height=600 width=900 src="https://vitalflux.com/wp-content/uploads/2020/07/Screenshot-2020-07-09-at-10.19.39-AM.png" />

```python
model_C = SVC(C=5)
model_C.fit(X_train, y_train)


print(model_C.score(X_test, y_test))
print(model_C.score(X_train, y_train))
```

```python
model_C = SVC(C=0.01) #hypermeter
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
print(model_C.score(X_train, y_train))
```

# **2. Gamma**

<img height=1000 width=1300 src="https://towardsdatascience.com/wp-content/uploads/2018/12/1B8e0TE2rTx8gdOH1FA1rXg.png" />

```python
model_g = SVC(gamma=1)
model_g.fit(X_train, y_train)
```

```python
print(model_g.score(X_test, y_test))
print(model_g.score(X_train, y_train))
```

# **3. Kernel**

<img height=800 width=900 src="https://miro.medium.com/v2/resize:fit:800/1*zqEbdsAbN4QqDdJ6L5pjnw.png" />

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
print(model_linear_kernal.score(X_test, y_test))
print(model_linear_kernal.score(X_train, y_train))
```

```python
model_linear_rbf = SVC(gamma=1,kernel='rbf')
model_linear_rbf.fit(X_train, y_train)
```

```python
print(model_linear_rbf.score(X_test, y_test))
print(model_linear_rbf.score(X_train, y_train))
```
