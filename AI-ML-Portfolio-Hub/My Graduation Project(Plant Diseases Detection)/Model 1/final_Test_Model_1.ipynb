### ***🎯 Custom F1 Score Metric Implementation***

```python
class F1Score(Metric):
    def __init__(self, name='f1_score', threshold=0.5, **kwargs):
        super(F1Score, self).__init__(name=name, **kwargs)
        self.threshold = threshold
        self.tp = self.add_weight(name='tp', initializer='zeros')
        self.fp = self.add_weight(name='fp', initializer='zeros')
        self.fn = self.add_weight(name='fn', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        y_pred = K.cast(y_pred > self.threshold, 'float32')
        tp = K.sum(y_true * y_pred)
        fp = K.sum((1 - y_true) * y_pred)
        fn = K.sum(y_true * (1 - y_pred))

        self.tp.assign_add(tp)
        self.fp.assign_add(fp)
        self.fn.assign_add(fn)

    def result(self):
        precision = self.tp / (self.tp + self.fp + K.epsilon())
        recall = self.tp / (self.tp + self.fn + K.epsilon())
        return 2 * (precision * recall) / (precision + recall + K.epsilon())

    def reset_states(self):
        self.tp.assign(0)
        self.fp.assign(0)
        self.fn.assign(0)
```

### ***🧪⚙️ PROFESSIONAL TEST DATA PREPARATION ⚙️🧪***

```python
def create_balanced_test_set_from_invalid_ratio(
    plantvillage_dir,
    plantdoc_train_dir,
    plantdoc_test_dir,
    invalid_dir,
    test_dir,
    invalid_percent=0.15,
    seed=42
):

    random.seed(seed)

    # Step 1: Collect all image paths
    def collect_images(directory):
        image_paths = []
        for root, _, files in os.walk(directory):
            for file in files:
                if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                    image_paths.append(os.path.join(root, file))
        return image_paths

    plantdoc_test_images = collect_images(plantdoc_test_dir)
    plantvillage_images = collect_images(plantvillage_dir)
    plantdoc_train_images = collect_images(plantdoc_train_dir)
    invalid_images = collect_images(invalid_dir)

    # Step 2: Determine number of invalid test images (15% of all invalids)
    num_invalid_test = max(1, int(len(invalid_images) * invalid_percent))
    invalid_test_images = random.sample(invalid_images, num_invalid_test)

    # Step 3: Start with all PlantDoc test images as valid
    valid_images = list(plantdoc_test_images)

    # Step 4: If needed, sample more valid images from PV + PlantDoc/train
    if len(valid_images) < num_invalid_test:
        extra_needed = num_invalid_test - len(valid_images)
        extra_pool = plantvillage_images + plantdoc_train_images
        extra_valid_images = random.sample(extra_pool, min(extra_needed, len(extra_pool)))
        valid_images += extra_valid_images

    # Trim in case valid exceeds invalid (optional safety)
    valid_images = valid_images[:num_invalid_test]

    # Step 5: Setup test set directories
    valid_test_dir = os.path.join(test_dir, "valid")
    invalid_test_dir = os.path.join(test_dir, "invalid")
    os.makedirs(valid_test_dir, exist_ok=True)
    os.makedirs(invalid_test_dir, exist_ok=True)

    # Step 6: Copy valid images
    for i, path in enumerate(valid_images):
        ext = os.path.splitext(path)[1]
        shutil.copy(path, os.path.join(valid_test_dir, f"valid_{i}{ext}"))

    # Step 7: Copy invalid images
    for i, path in enumerate(invalid_test_images):
        ext = os.path.splitext(path)[1]
        shutil.copy(path, os.path.join(invalid_test_dir, f"invalid_{i}{ext}"))

    print(f"✅ Created balanced test set with {len(valid_images)} valid and {len(invalid_test_images)} invalid images.")
```

### ***🎨 Image Preprocessing Function with Data Augmentation 🎨***

```python
from tensorflow.keras.applications.resnet_v2 import preprocess_input as resnet_preprocess

IMG_SIZE = (224, 224)

def preprocess_image(image, label, is_train=True):
    image = tf.image.resize(image, IMG_SIZE)
    image = tf.image.convert_image_dtype(image, tf.float32)
    if is_train:
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_brightness(image, 0.2)
        image = tf.image.random_contrast(image, 0.8, 1.2)
        image = tf.image.random_saturation(image, 0.8, 1.2)
        image = tf.image.random_hue(image, 0.05)
    image = resnet_preprocess(image)
    return image, label
```

### ***📊 Professional Model Evaluation with Metrics & Visualizations 📈***

```python
def evaluate_model_professionally(model, test_dir, img_size=(224, 224), batch_size=32, models_dir=None):
    """
    Comprehensive evaluation of the model with multiple metrics and visualizations
    """
    # Create test dataset
    test_dataset = tf.keras.utils.image_dataset_from_directory(
        test_dir,
        labels='inferred',
        label_mode='binary',
        image_size=img_size,
        batch_size=batch_size,
        shuffle=False
    )

    # Preprocess images (same as training)
    test_dataset = test_dataset.map(
        lambda x, y: preprocess_image(x, y, is_train=False),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    # Get true labels and predictions
    y_true = np.concatenate([y for x, y in test_dataset], axis=0)
    y_pred = model.predict(test_dataset)
    y_pred_classes = (y_pred > 0.5).astype(int)

    # Classification report
    print("\n" + "="*50)
    print("CLASSIFICATION REPORT")
    print("="*50)
    print(classification_report(y_true, y_pred_classes,
                              target_names=['Invalid', 'Valid']))

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Invalid', 'Valid'],
                yticklabels=['Invalid', 'Valid'])
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    if models_dir:
        plt.savefig(os.path.join(models_dir, 'confusion_matrix.png'))
    plt.show()

    # ROC Curve
    fpr, tpr, thresholds = roc_curve(y_true, y_pred)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2,
             label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    if models_dir:
        plt.savefig(os.path.join(models_dir, 'roc_curve.png'))
    plt.show()

    # Per-class metrics
    tn, fp, fn, tp = cm.ravel()
    metrics = {
        'Accuracy': (tp + tn) / (tp + tn + fp + fn),
        'Precision': tp / (tp + fp),
        'Recall': tp / (tp + fn),
        'Specificity': tn / (tn + fp),
        'F1-Score': 2 * tp / (2 * tp + fp + fn),
        'AUC': roc_auc
    }

    print("\n" + "="*50)
    print("DETAILED METRICS")
    print("="*50)
    for name, value in metrics.items():
        print(f"{name}: {value:.4f}")

    return metrics
```

### ***🔥 Grad-CAM Visualization for Model Interpretability 🔥***

```python
def show_gradcam(model, img, label, layer_name=None):
    """
    Generates Grad-CAM visualization for a given image and model
    """
    try:
        # Remove batch dimension if present
        if len(img.shape) == 4:
            img = img[0]

        # Convert image to array and add batch dimension
        img_array = tf.expand_dims(img, axis=0)

        # First ensure model is built by making a dummy prediction
        if not model.built:
            model.predict(img_array)

        # Find the last convolutional layer
        if layer_name is None:
            for layer in reversed(model.layers):
                # Check for Conv2D or similar layers
                if hasattr(layer, 'kernel') and len(layer.weights) > 0:
                    try:
                        if len(layer.output.shape) == 4:  # 4D output (batch, h, w, channels)
                            layer_name = layer.name
                            break
                    except:
                        continue

        if layer_name is None:
            print("Could not find suitable convolutional layer. Showing original image.")
            plt.imshow(img.numpy().astype('uint8'))
            plt.axis('off')
            return

        # Create gradient model
        grad_model = tf.keras.models.Model(
            inputs=[model.inputs],
            outputs=[model.get_layer(layer_name).output, model.output]
        )

        # Compute gradients
        with tf.GradientTape() as tape:
            conv_outputs, predictions = grad_model(img_array)
            loss = predictions[:, 0] if label == 1 else 1 - predictions[:, 0]

        # Get gradients
        grads = tape.gradient(loss, conv_outputs)
        if grads is None:
            print("Gradients are None. Showing original image.")
            plt.imshow(img.numpy().astype('uint8'))
            plt.axis('off')
            return

        pooled_grads = tf.reduce_mean(grads[0], axis=(0, 1))
        conv_output = conv_outputs[0]

        # Weight the output feature map
        for i in range(pooled_grads.shape[0]):
            conv_output[:, :, i] *= pooled_grads[i]

        # Create heatmap
        heatmap = tf.reduce_mean(conv_output, axis=-1)
        heatmap = np.maximum(heatmap, 0)
        heatmap /= tf.reduce_max(heatmap) + 1e-10  # Avoid division by zero
        heatmap = heatmap.numpy()

        # Resize and apply colormap
        heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
        heatmap = np.uint8(255 * heatmap)
        heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

        # Superimpose on original image
        superimposed_img = cv2.addWeighted(
            cv2.cvtColor(img.numpy().astype('uint8'), cv2.COLOR_RGB2BGR),
            0.6,
            heatmap,
            0.4,
            0
        )

        plt.imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
        plt.axis('off')

    except Exception as e:
        print(f"Error generating Grad-CAM: {str(e)}")
        plt.imshow(img.numpy().astype('uint8'))
        plt.axis('off')
```

### ***🖼️🔍 Visualize Predictions with Grad-CAM Explanations 🔥***

```python
def visualize_predictions(model, test_dir, num_samples=5, img_size=(224, 224), models_dir=None):
    """
    Visualizes sample predictions with Grad-CAM explanations
    """
    # Load a few samples
    test_dataset = tf.keras.utils.image_dataset_from_directory(
        test_dir,
        labels='inferred',
        label_mode='binary',
        image_size=img_size,
        batch_size=num_samples,
        shuffle=True
    )

    # Get a batch
    for images, labels in test_dataset.take(1):
        # Make predictions
        preds = model.predict(images)

        # Visualize each sample
        plt.figure(figsize=(15, 3*num_samples))
        for i in range(num_samples):
            # Original image
            plt.subplot(num_samples, 2, 2*i+1)
            plt.imshow(images[i].numpy().astype('uint8'))
            plt.title(f"True: {'Valid' if labels[i] else 'Invalid'}\n"
                     f"Pred: {'Valid' if preds[i] > 0.5 else 'Invalid'} ({preds[i][0]:.2f})")
            plt.axis('off')

            # Grad-CAM
            plt.subplot(num_samples, 2, 2*i+2)
            show_gradcam(model, images[i], labels[i])

        plt.tight_layout()
        if models_dir:
            plt.savefig(os.path.join(models_dir, 'sample_predictions.png'))
        plt.show()
```
