ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸ðŸŒ¸ âœ¿ âœ¿ âœ¿ âœ¿ âœ¿ ðŸŒ¸


```python
MODEL1_SIZE = (224, 224)
MODEL234_SIZE = (384, 384)
```

ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿


```python
# ----------- ENHANCEMENT STEP BEFORE MODEL 1 -----------
def enhance_if_blurry(img_path, threshold=100.0):
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur_score = cv2.Laplacian(gray, cv2.CV_64F).var()

    if blur_score < threshold:
        print(f"[INFO] Image is blurry (score={blur_score:.2f})... enhancing contrast.")
        lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
        l, a, b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
        cl = clahe.apply(l)
        enhanced_img = cv2.merge((cl, a, b))
        final_img = cv2.cvtColor(enhanced_img, cv2.COLOR_LAB2BGR)
        final_rgb = cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB)


        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(img_rgb)
        plt.title("Before Enhancement")
        plt.axis("off")

        plt.subplot(1, 2, 2)
        plt.imshow(final_rgb)
        plt.title("After Enhancement")
        plt.axis("off")
        plt.tight_layout()
        plt.show()

        temp_path = "/tmp/enhanced_temp.jpg"
        cv2.imwrite(temp_path, final_img)
        return temp_path
    else:
        print(f"[INFO] Image is clear (score={blur_score:.2f}). No enhancement needed.")
        print("The image is ready for Processing")


        plt.imshow(img_rgb)
        plt.title("Clear Image (No Enhancement)")
        plt.axis("off")
        plt.show()

        return img_path
```

ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿ðŸŒ¿ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ ðŸŒ¿


```python
def preprocess_for_model1(image_path):
    img = keras_image.load_img(image_path, target_size=MODEL1_SIZE)
    img_array = keras_image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = resnet_preprocess(img_array)
    return img_array, img
```

```python
def preprocess_for_model234(image_path):
    img = keras_image.load_img(image_path, target_size=MODEL234_SIZE)
    img_array = keras_image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = effnet_preprocess(img_array)
    return img_array
```

```python
# ----------- MAIN PREDICTION PIPELINE -----------

def predict_pipeline(img_path, class_names_model3, class_names_model4):
    # Step 1 - Enhance if blurry
    enhanced_img_path = enhance_if_blurry(img_path)

    # Step 1 - Validity Check
    img_model1_array, raw_img = preprocess_for_model1(enhanced_img_path)
    pred1 = model1.predict(img_model1_array)[0][0]
    validity = "Valid" if pred1 > 0.5 else "Invalid"
    confidence1 = pred1 if pred1 > 0.5 else 1 - pred1

    if validity == "Invalid":
        return {
            "validity": (validity, confidence1),
            "health": None,
            "plant_type": None,
            "disease": None,
            "raw_img": raw_img
        }

    # Preprocess for Models 2, 3, 4
    img_array_384 = preprocess_for_model234(img_path)

    # Step 2 - Health Status
    pred2 = model2.predict(img_array_384)[0][0]
    health = "Infected" if pred2 > 0.5 else "Healthy"
    confidence2 = pred2 if pred2 > 0.5 else 1 - pred2

    # Step 3 - Plant Type
    pred3 = model3.predict(img_array_384)[0]
    class_idx3 = np.argmax(pred3)
    confidence3 = pred3[class_idx3]
    plant_type = class_names_model3[class_idx3]

    # Step 4 - Disease Type (only if infected)
    disease = None
    confidence4 = None
    if health == "Infected":
        pred4 = model4.predict(img_array_384)[0]
        class_idx4 = np.argmax(pred4)
        confidence4 = pred4[class_idx4]
        disease = class_names_model4[class_idx4]

    return {
        "validity": (validity, confidence1),
        "health": (health, confidence2),
        "plant_type": (plant_type, confidence3),
        "disease": (disease, confidence4),
        "raw_img": raw_img
    }
```

```python
# ============ DISPLAY RESULTS ============
def print_results(results):
    print("\n=== Plant Disease Detection Results ===")

    validity, conf1 = results["validity"]
    print(f"1. Validity Check: {validity} ({conf1:.2%})")

    plt.figure(figsize=(6, 6))
    plt.imshow(results["raw_img"])
    plt.title(f"Input Image - Validity: {validity} ({conf1:.2%})")
    plt.axis("off")
    plt.show()

    if validity == "Invalid":
        print("Image is invalid for plant disease detection. Please try another image.")
        return

    health, conf2 = results["health"]
    print(f"2. Health Status: {health} ({conf2:.2%})")

    plant_type, conf3 = results["plant_type"]
    print(f"3. Plant Type: {plant_type} ({conf3:.2%})")

    disease_info = results.get("disease")
    if disease_info and disease_info[0] is not None:
        disease, conf4 = disease_info
        print(f"4. Disease Type: {disease} ({conf4:.2%})")

        recs = get_disease_recommendations(disease)
        print("\nTreatment Recommendations:")
        for i, rec in enumerate(recs, 1):
            print(f"  {i}. {rec}")
    else:
        print("4. Disease Type: None (Plant is healthy or no disease detected)")
```
