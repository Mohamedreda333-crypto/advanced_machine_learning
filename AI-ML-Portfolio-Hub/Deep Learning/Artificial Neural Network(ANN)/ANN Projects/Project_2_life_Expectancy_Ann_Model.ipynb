```python
features_to_drop = [
    "Country",
    "Year",
    "percentage expenditure",
    "under-five deaths",
    "thinness 5-9 years",
    "Diphtheria"
]

df_fs = df.drop(columns=features_to_drop)
```

```python
df_fs.head()
```

<span style="color:white; background-color:red; font-size:20px; font-weight:bold">
Check Data types 
</span>

```python
dtype = df_fs.dtypes
N_unique = df_fs.nunique()
pd.DataFrame({"Number of Dtyes of each column" : dtype , "Number of unique" : N_unique}).T
```

<span style="color:white; background-color:red; font-size:20px; font-weight:bold">
 Handle Null Values 
</span>

```python
null = df_fs.isnull().sum()
ratio = null / df_fs.shape[0]
pd.DataFrame({"Number of Nulls Values" : null , "ratio" : ratio * 100}).T
```

```python
df_fs = df_fs.dropna(subset=["Life expectancy"])
```

<span style="color:white; background-color:red; font-size:20px; font-weight:bold">
Categorical Feature (Status)</span>

```python
df_fs["Status"] = df_fs["Status"].map({
    "Developed": 1,
    "Developing": 0
})
```

<span style="color:white; background-color:red; font-size:20px; font-weight:bold">
 Numerical Features â€“ Smart Imputation
</span>

<span style="color:black; background-color:yellow; font-size:20px; font-weight:bold">
Impute using Median grouped by Status because Developed countries â‰  developing countries
</span>

```python
num_cols = df_fs.select_dtypes(include="number").columns

for col in num_cols:
    df_fs[col] = df_fs[col].fillna(
        df_fs.groupby("Status")[col].transform("median")
    )
```

```python
df_fs.isnull().sum()
```

<span style="color:white; background-color:red; font-size:20px; font-weight:bold">
Data splitting 
</span>

```python
X = df_fs.drop("Life expectancy", axis=1)
y = df_fs["Life expectancy"]
```

```python
X
```

```python
y
```

<span style="color:white; background-color:red; font-size:20px; font-weight:bold">
Normalization(Scalling) 
</span>

```python
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42
)
```

```python
print(f"Shape of X_train is: {X_train.shape}")
print(f"Shape of Y_train is: {y_train.shape}\n")
print(f"Shape of X_test is: {X_test.shape}")
print(f"Shape of Y_test is: {y_test.shape}")
```

<div style="text-align:center;">
<span style="color:white; background-color:darkblue; font-size:20px; font-weight:bold; padding:7px; border-radius:5px;">
ANN Models
</span>
</div>

```python
ann = Sequential([
    Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.3),

    Dense(64, activation="relu"),
    Dropout(0.2),

    Dense(64, activation="relu"),
    Dense(1)
])
```

```python
ann.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error','mean_squared_error'])
```

```python
ann.summary()
```

```python
history = ann.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    verbose=1
)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_pred_ann = ann.predict(X_test).flatten()

ann_r2 = r2_score(y_test, y_pred_ann)
```

```python
ann_r2 
```

```python
ann_mae = mean_absolute_error(y_test, y_pred_ann)
ann_mae
```

## Conclusion

In this project, multiple machine learning and deep learning models were evaluated to predict life expectancy based on health, economic, and social indicators.

Despite implementing a well-structured Artificial Neural Network (ANN) with proper scaling and regularization, tree-based modelsâ€”particularly **XGBoost**â€”achieved superior performance. This outcome highlights a key data science principle: **model effectiveness depends on data characteristics rather than model complexity**.

Given the structured and tabular nature of the dataset, along with a relatively limited number of samples, traditional machine learning models were better able to capture non-linear relationships and feature interactions, resulting in stronger generalization on unseen data.

This study reinforces the importance of informed model selection and demonstrates that deep learning is not always the optimal solution for every problem.


<span style="color:black; background-color:yellow; font-size:20px; font-weight:bold">
XGBoost outperforms other models because it combines gradient boosting with strong regularization, efficient optimization, and the ability to capture complex non-linear feature interactions, making it especially effective for tabular real-world data.</span>
